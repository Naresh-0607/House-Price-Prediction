```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
house_data = pd.read_csv("data.csv")
```

```python
house_filtered = house_data[(house_data['price'] < 150000) & (house_data['price'] != 0)]
# house_filtered = house_data[house_data['price'] != 0]
house_filtered = house_filtered[['price', 'sqft_living']]
house_filtered.to_csv("filtered_data.csv", index=False)
house_filtered
```

```python
df = pd.read_csv("filtered_data.csv")
```

```python
plt.scatter(df.sqft_living, df.price, color='red', marker='+')
plt.xlabel("Sqft")
plt.ylabel("Price")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['sqft_living']], df.price)
```

```python
reg.predict([[1750]])
```
